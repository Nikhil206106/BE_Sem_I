```python
import matplotlib.pyplot as plt
```

```python
# Define the cost function
def cost_function(x):
    return (x + 5) ** 2
```

```python
# Define the gradient of the function
def gradient(x):
    return 2 * (x + 5)
```

```python
# Hyperparameters
learning_rate = 0.1
initial_x = 3.0
num_iterations = 100
```

```python
# Lists to store values for plotting
x_values = []
y_values = []
```

```python
# Initialize x
x = initial_x
```

```python
# Gradient Descent loop
for i in range(num_iterations):
    x_values.append(x)
    y_values.append(cost_function(x))
    
    grad = gradient(x)
    x = x - learning_rate * grad  # Update x
    
    print(f"Iteration {i+1}: x = {x}, Cost = {cost_function(x)}")
```

```python
# Final results
print(f"\nOptimal x: {x}")
print(f"Minimum value of y: {cost_function(x)}")
```

```python
# Plotting the convergence
plt.plot(x_values, y_values, 'ro-')
plt.title('Gradient Descent Visualization for y = (x + 5)^2')
plt.xlabel('x')
plt.ylabel('y')
plt.show()
```
